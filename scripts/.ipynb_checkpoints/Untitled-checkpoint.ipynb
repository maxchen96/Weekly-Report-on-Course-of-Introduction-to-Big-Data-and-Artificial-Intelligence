```python
import matplotlib.pyplot as plt
import matplotlib
# 正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(38,8))
data = [80, 4, 3, 3, 2, 2, 2, 1, 1, 1, 1, 1, 1, 1, 1]
labels = ['材料科学与工程', '计算机与信息工程', '资源环境', '生命科学', '体育学', '哲学', '外国语', '化学生物学', '美术学', '化学工程与工艺', '艺术学院数字媒体艺术', '商学院', '国际事务与国际关系', '化学与化工', '数据科学与大数据技术']

plt.yticks(fontsize=20)
plt.xticks(fontsize=15)
plt.bar(range(len(data)), data, tick_label=labels, alpha=0.8)
plt.title('各专业学生选课人数',verticalalignment='bottom', fontsize=30)

for a,b in enumerate(data):
    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=20)
#横坐标label斜放
#ax = plt.gca()
#for label in ax.xaxis.get_ticklabels():
#    label.set_rotation(45)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

plt.rcParams['axes.labelsize']=16
plt.rcParams['xtick.labelsize']=14
plt.rcParams['ytick.labelsize']=14
plt.rcParams['legend.fontsize']=12
plt.rcParams['figure.figsize']=[15,15]
plt.style.use("ggplot")
price = [72,32]

fig,ax = plt.subplots(figsize=(12,3),dpi=80)

x = range(len(price))
# 添加刻度标签
labels = np.array(['男','女'])
#在子图对象上画条形图，并添加x轴标签，图形的主标题
ax.barh(x,price,tick_label=labels,alpha = 0.8)

ax.set_xlabel('人数',color='k')
ax.set_title('男女学生选课人数')

# 设置Y轴的刻度范围
ax.set_xlim([0,80])

# 为每个条形图添加数值标签
for x,y in enumerate(price):
    ax.text(y+0.2,x,y,va='center',fontsize=14)
    
# 显示图形   
plt.show()
```
