```python
import matplotlib.pyplot as plt
import numpy as np
# 正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

plt.rcParams['axes.labelsize']=16
plt.rcParams['xtick.labelsize']=14
plt.rcParams['ytick.labelsize']=14
plt.rcParams['legend.fontsize']=12
plt.rcParams['figure.figsize']=[15,15]
plt.style.use("ggplot")
price = [72,32]

fig,ax = plt.subplots(figsize=(12,3),dpi=80)

x = range(len(price))
# 添加刻度标签
labels = np.array(['男','女'])
#在子图对象上画条形图，并添加x轴标签，图形的主标题
ax.barh(x,price,tick_label=labels,alpha = 0.8)

ax.set_xlabel('人数',color='k')
ax.set_title('男女学生选课人数')

# 设置Y轴的刻度范围
ax.set_xlim([0,80])

# 为每个条形图添加数值标签
for x,y in enumerate(price):
    ax.text(y+0.2,x,y,va='center',fontsize=14)
    
# 显示图形   
plt.show()
```

```python
import numpy as np
import pandas as pd

pd.options.display.max_rows = 40
tips = pd.read_csv("001.csv")
tips2 = pd.read_csv("003.csv")
tipsall = pd.read_csv("all.csv")

print('课程总人数：')
print(tipsall['actor'].drop_duplicates().count())
print('------------------------------------------------------------')
print('用户列表：')
print(tipsall.groupby('actor')['actor'].count())
print('============================================================')
print('看课总人数：')
print(tips2['actor'].drop_duplicates().count())
print('------------------------------------------------------------')
print('用户列表：')
print(tips2.groupby('actor')['actor'].count())
print('============================================================')
print('看视频总人数：')
print(tips['actor'].drop_duplicates().count())
print('------------------------------------------------------------')
print('用户列表：')
print(tips.groupby('actor')['actor'].count())

print('============================================================')
print('用户列表及每小部分观看时长：')
print(tips.groupby(['actor', 'object'])['metadata'].sum().head(264))
tips.replace('learn-unit:大数据与人工智能通识导论-05 数据思维-视频5-1：开篇实例', '5-1：开篇实例', inplace=True)
tips.replace('learn-unit:大数据与人工智能通识导论-05 数据思维-视频5-2：问题求解与思维方式', '5-2：问题求解与思维方式', inplace=True)
tips.replace('learn-unit:大数据与人工智能通识导论-05 数据思维-视频5-3：计算思维与数据思维', '5-3：计算思维与数据思维', inplace=True)
tips.replace('learn-unit:大数据与人工智能通识导论-05 数据思维-视频5-4：像数据科学家一样思考', '5-4：像数据科学家一样思考', inplace=True)
tips.replace('learn-unit:大数据与人工智能通识导论-05 数据思维-视频5-5：总结', '5-5：总结', inplace=True)

temp = tips.groupby(['actor', 'object'])['metadata'].sum()
print('============================================================')
print('个各视频平均观看时长：')
print(tips.groupby('object')['metadata'].mean().head(264))
temp2 = tips.groupby('object')['metadata'].mean().head(264)

temp.to_csv("data.csv",header = True)
temp2.to_csv("data2.csv",header = True)


print('============================================================')
print('各视频观看列表：')
print(tips.groupby(['object','actor'])['metadata'].sum().head(200))
temp3 = tips.groupby(['object','actor'])['metadata'].sum().head(200)
temp3.to_csv("data3.csv",header = True)
```
