```python
import matplotlib.pyplot as plt
import matplotlib
# 
plt.rcParams['font.sans-serif']=['SimHei']
# 
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(38,8))
data = [80, 4, 3, 3, 2, 2, 2, 1, 1, 1, 1, 1, 1, 1, 1]
labels = ['', '', '', '', '', '', '', '', '', '', '', '', '', '', '']

plt.yticks(fontsize=20)
plt.xticks(fontsize=15)
plt.bar(range(len(data)), data, tick_label=labels, alpha=0.8)
plt.title('',verticalalignment='bottom', fontsize=30)

for a,b in enumerate(data):
    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=20)
#label
#ax = plt.gca()
#for label in ax.xaxis.get_ticklabels():
#    label.set_rotation(45)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 
plt.rcParams['font.sans-serif']=['SimHei']
# 
plt.rcParams['axes.unicode_minus']=False

plt.rcParams['axes.labelsize']=16
plt.rcParams['xtick.labelsize']=14
plt.rcParams['ytick.labelsize']=14
plt.rcParams['legend.fontsize']=12
plt.rcParams['figure.figsize']=[15,15]
plt.style.use("ggplot")
price = [72,32]

fig,ax = plt.subplots(figsize=(12,3),dpi=80)

x = range(len(price))
# 
labels = np.array(['',''])
#x
ax.barh(x,price,tick_label=labels,alpha = 0.8)

ax.set_xlabel('',color='k')
ax.set_title('')

# Y
ax.set_xlim([0,80])

# 
for x,y in enumerate(price):
    ax.text(y+0.2,x,y,va='center',fontsize=14)
    
#    
plt.show()
```
